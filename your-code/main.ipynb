```python
# Your code here:
print("Dataset shape:", data.shape)
print("\nDataset info:")
data.info()
print("\nMissing values per column:")
print(data.isnull().sum())

# Descriptive statistics
print("\nDescriptive statistics:")
data_stats = data.describe()
print(data_stats)

# Check distribution of categorical variables
print("\nDistribution of Channel:")
print(data['Channel'].value_counts())
print("\nDistribution of Region:")
print(data['Region'].value_counts())

# Let's visualize the data distribution for each product category
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure for histograms of spending categories
plt.figure(figsize=(15, 10))
for i, feature in enumerate(['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']):
    plt.subplot(2, 3, i+1)
    sns.histplot(data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.tight_layout()
plt.show()

# Let's check for correlations between variables
plt.figure(figsize=(10, 8))
correlation_matrix = data.iloc[:, 2:].corr()  # Excluding Channel and Region
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Let's check for outliers using box plots
plt.figure(figsize=(15, 8))
sns.boxplot(data=data.iloc[:, 2:])
plt.title('Box Plot of Spending Categories')
plt.xticks(rotation=45)
plt.show()

# Let's examine if the data exhibits the Pareto principle
# Sort customers by total spending
data['Total_Spending'] = data.iloc[:, 2:].sum(axis=1)
sorted_customers = data.sort_values('Total_Spending', ascending=False)

# Calculate cumulative percentage of total spending
total_spend = sorted_customers['Total_Spending'].sum()
sorted_customers['Cumulative_Percentage'] = 100 * sorted_customers['Total_Spending'].cumsum() / total_spend

# Plot Pareto curve
plt.figure(figsize=(10, 6))
plt.plot(100 * np.arange(1, len(sorted_customers) + 1) / len(sorted_customers), 
         sorted_customers['Cumulative_Percentage'], 
         marker='.', linestyle='-')
plt.plot([0, 100], [0, 100], 'k--')  # Diagonal line
plt.plot([20, 20], [0, 100], 'r--')  # 20% customers vertical line
plt.plot([0, 20], [80, 80], 'r--')   # 80% spending horizontal line
plt.xlabel('Cumulative Percentage of Customers')
plt.ylabel('Cumulative Percentage of Spending')
plt.title('Pareto Analysis of Customer Spending')
plt.grid(True)
plt.show()

# Calculate what percentage of spending comes from top 20% customers
top_20_percent = int(len(sorted_customers) * 0.2)
top_20_spending = sorted_customers.iloc[:top_20_percent]['Total_Spending'].sum()
pareto_percentage = 100 * top_20_spending / total_spend
print(f"\nTop 20% of customers account for {pareto_percentage:.2f}% of total spending")

# Let's also examine differences between customer channels (Horeca vs Retail)
plt.figure(figsize=(15, 10))
for i, feature in enumerate(['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='Channel', y=feature, data=data)
    plt.title(f'{feature} by Channel')
plt.tight_layout()
plt.show()
```

# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
# Your code here
# Check for outliers using the IQR method to detect extreme values
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers.index

# Check outliers in each spending category
spending_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
all_outliers = set()

for col in spending_cols:
    outliers = detect_outliers(data, col)
    print(f"Outliers in {col}: {len(outliers)}")
    all_outliers.update(outliers)

print(f"\nTotal unique outlier rows: {len(all_outliers)} out of {len(data)} ({100*len(all_outliers)/len(data):.2f}%)")

# Create a clean dataset without outliers
clean_customers = data.drop(list(all_outliers), axis=0).reset_index(drop=True)
print(f"\nClean dataset shape: {clean_customers.shape}")

# Check the distribution of data after removing outliers
plt.figure(figsize=(15, 10))
for i, feature in enumerate(spending_cols):
    plt.subplot(2, 3, i+1)
    sns.histplot(clean_customers[feature], kde=True)
    plt.title(f'Distribution of {feature} (after outlier removal)')
    plt.tight_layout()
plt.show()

# Let's also apply log transformation to address the skewness
# We'll add a small constant (1) to handle zero values
clean_customers_log = clean_customers.copy()
for col in spending_cols:
    clean_customers_log[col] = np.log1p(clean_customers_log[col])  # log1p = log(1+x)

# Plot the log-transformed distributions
plt.figure(figsize=(15, 10))
for i, feature in enumerate(spending_cols):
    plt.subplot(2, 3, i+1)
    sns.histplot(clean_customers_log[feature], kde=True)
    plt.title(f'Log-transformed Distribution of {feature}')
    plt.tight_layout()
plt.show()

# Summary of cleaning and transformation
print("\nData cleaning summary:")
print(f"1. Removed {len(all_outliers)} outlier rows")
print(f"2. Applied log transformation to address skewness")
print(f"3. Original data shape: {data.shape}, Clean data shape: {clean_customers.shape}")
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Your code here:
# Identify which columns need scaling
# Only scale the numeric spending columns, not the categorical columns (Channel, Region)
spending_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Create a StandardScaler instance
scaler = StandardScaler()

# Let's scale both versions of our data:
# 1. For clean_customers (outliers removed)
customers_scaled = clean_customers.copy()
customers_scaled[spending_cols] = scaler.fit_transform(clean_customers[spending_cols])

# 2. For clean_customers_log (outliers removed and log-transformed)
customers_log_scaled = clean_customers_log.copy()
customers_log_scaled[spending_cols] = scaler.fit_transform(clean_customers_log[spending_cols])

# Let's look at the statistics of the scaled data
print("Statistics of scaled data (outliers removed):")
print(customers_scaled[spending_cols].describe().round(2))

# Visualize the scaled distributions
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))
for i, feature in enumerate(spending_cols):
    plt.subplot(2, 3, i+1)
    sns.histplot(customers_scaled[feature], kde=True)
    plt.title(f'Scaled Distribution of {feature}')
    plt.axvline(x=0, color='r', linestyle='--')  # Line at mean (0)
    plt.axvline(x=1, color='g', linestyle='--')  # Line at 1 std dev
    plt.axvline(x=-1, color='g', linestyle='--') # Line at -1 std dev
    plt.tight_layout()
plt.show()

# Let's also check the correlation matrix of scaled data to see if scaling affected relationships
plt.figure(figsize=(10, 8))
correlation_matrix = customers_scaled[spending_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix (after scaling)')
plt.show()

# For the final step, create a copy of the scaled data that we'll use for clustering
# We'll use the version with outliers removed and standard scaling
customers_scale = customers_scaled[spending_cols].copy()

print("\nFinal data shape for clustering:", customers_scale.shape)
print("Mean values of scaled features should be close to 0:")
print(customers_scale.mean().round(4))
print("\nStandard deviation of scaled features should be close to 1:")
print(customers_scale.std().round(4))
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# First, we need to determine the optimal number of clusters
# We'll use the Elbow Method to find the best value for k
# The Elbow Method plots the explained variance vs. number of clusters

# Initialize an empty list to store the inertia (within-cluster sum of squares)
inertia = []
k_range = range(1, 11)  # Testing K from 1 to 10

# Compute inertia for different values of k
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_scale)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method results
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-cluster sum of squares)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()

k_optimal = 3

# Fit K-means with the optimal number of clusters
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
kmeans.fit(customers_scale)

# Get the cluster labels
labels = kmeans.labels_

# Add the cluster labels to the original cleaned dataframe
clean_customers['KMeans_Cluster'] = labels

# Let's examine the cluster sizes
print("Cluster sizes:")
print(pd.Series(labels).value_counts())

# Let's look at the cluster centroids
centroids = pd.DataFrame(kmeans.cluster_centers_, 
                         columns=customers_scale.columns,
                         index=[f'Cluster {i}' for i in range(k_optimal)])
print("\nCluster centroids:")
print(centroids)

# Visualize the clusters in 2D using the first two principal components
from sklearn.decomposition import PCA

# Apply PCA to reduce the data to 2 dimensions
pca = PCA(n_components=2)
pca_result = pca.fit_transform(customers_scale)

# Create a dataframe with the PCA results and cluster labels
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['Cluster'] = labels

# Plot the clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis', s=100, alpha=0.7)

# Plot the centroids in the PCA space
centroids_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], s=300, c='red', marker='X', label='Centroids')

plt.title(f'K-means Clustering with {k_optimal} Clusters (PCA visualization)')
plt.legend()
plt.grid(True)
plt.show()

# Let's also analyze the characteristics of each cluster by looking at the average spending
cluster_means = clean_customers.groupby('KMeans_Cluster')[spending_cols].mean()
print("\nMean spending by cluster:")
print(cluster_means)

# Visualize the cluster means with a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cluster_means, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Average Spending by Cluster')
plt.show()
```
